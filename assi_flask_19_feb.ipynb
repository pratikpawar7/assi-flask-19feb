```python
'''
Q3. What is App routing in Flask? Why do we use app routes?
'''
App Routing :
    App Routing means mapping the URLs to a specific function that will handle the logic for that URL. 
    Modern web frameworks use more meaningful URLs to help users remember the URLs and make navigation simpler.
    
use of app routing :
    
    App routing is used to manage navigation and URLs in web applications.

    It helps to organize the application into different pages or views, making it easier to navigate and maintain.

    Routing also enables the user to bookmark or share a specific page or view of the application.
```

```python
'''
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
''' 

''' 
URL Building:   url_for()  function 

To build a URL to a specific function, use the url_for() function. 
It accepts the name of the function as its first argument and any number of keyword arguments, 
each corresponding to a variable part of the URL rule'''


from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Welcome to the homepage!'

@app.route('/about')
def about():
    return 'This is the about page.'

@app.route('/user/<username>')
def user_profile(username):
    return f'Hello, {username}!'

with app.test_request_context():
    print(url_for('index'))  # Output: /
    print(url_for('about'))  # Output: /about
    print(url_for('user_profile', username='John Doe'))  # Output: /user/John%20Doe

```
